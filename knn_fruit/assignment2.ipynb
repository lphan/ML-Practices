**1) First Step: Data Preparation**

(a) & (b) Importing data and deleting duplicates

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import sklearn.linear_model as sklm

data = pd.read_csv("assignment02.csv")
#getting the values from the data
datarray = data.values

#loop to go compare values within data
for i in range (len(datarray)):
    for j in range (i+1,len(datarray)-1):
        #comparing each row
        check = np.array_equal(datarray[i], datarray[j])
        if (check == True):
            #deleting equal rows
            datarray = np.delete(datarray,j,0)

#print(len(datarray))
print("Values without doubled values:")
print(datarray)
print(type(datarray))
```

(c) Replacing missing numerical values

```python
#loop to go compare values within data

#checking for radius
sum = 0
n= 0
for i in range (len(datarray)):
    if (datarray[i][1] == 0):
        fruitclass = datarray[i][3]
        print(fruitclass)
        
        #loop for getting the mean value       
        for j in range (len(datarray)):
            if(datarray[j][3] == fruitclass):  #check index
                sum = sum + datarray[j][1]
                n = n + 1
        print (sum)
        print (n)
        mean  = sum / (n-1)
        print (mean)
        datarray[i][1] = mean
        print(datarray[i])

# checking for weight
n=0
sum = 0
for i in range (len(datarray)):
    if (datarray[i][2] == 0):
        fruitclass = datarray[i][3]
        print(fruitclass)
        
        # loop for getting the mean value       
        for j in range (len(datarray)):
            if(datarray[j][3] == fruitclass):   #check for the index
                sum = sum + datarray[j][2]
                n = n + 1
        print (sum)
        print (n)
        mean  = sum / (n-1)
        print (mean)
        datarray[i][2] = mean
        print(datarray[i])

print(datarray)
```

(d) Transforming nominal features into numbers

```python
#adding 3 extra columns for the colors
new_datarray = np.zeros((len(datarray),3), dtype=int)
new_datarray = np.append(datarray,new_datarray,axis = 1)

#assigning values of colors in numerical values

for i in range (len(new_datarray)):
    if(new_datarray [i][0] == "Yellow"):
        new_datarray[i][4] = 1
    elif (new_datarray [i][0] == "Green"):
        new_datarray[i][5] = 1
    elif (new_datarray [i][0] == "Red"):
        new_datarray[i][6] = 1
# removing column with color nominal feature
new_datarray = np.delete(new_datarray,0,1)
print(new_datarray)
```

(e) NORMALIZATION OF OBSERVATIONS

```python
# getting min and max values in radius and weight columns
radius = []
for i in range (len(new_datarray)):
    radius.append(new_datarray[i][0])

weight = []
for i in range (len(new_datarray)):
    weight.append(new_datarray[i][1])

# normalizing the observations of each feature

norm_datarray = np.zeros((len(datarray),2), dtype=int)
norm_datarray = np.append(new_datarray,norm_datarray,axis = 1)
for i in range (len(norm_datarray)):
    norm_datarray[i][6] = (norm_datarray[i][0] - min(radius))/(max(radius)-min(radius))
    norm_datarray[i][7] = (norm_datarray[i][1] - min(weight))/(max(weight)-min(weight))
#    np.around(norm_datarray[i][6], decimals = 4)
#    np.around(norm_datarray[i][7], decimals = 4)
print(new_datarray)
print("******************")
print(norm_datarray)


```
