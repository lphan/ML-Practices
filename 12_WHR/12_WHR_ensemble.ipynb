```python
# create new feature (column) 'Happiness' from feature 'Happiness Score' 
values = [1 if h > 5 else 0 for h in train_data['Happiness Score'].values]
train_data['Happiness'] = pd.Series(values)
train_data.head()
```

```python
train_data.tail()
```

```python
cls_data = train_data.iloc[:, 3:13]
cls_data.head()
```

```python
# Ensemble Random Forest
clf_rf, y_test, y_predict = StartModSKL.classification_rf(cls_data, dependent_label='Happiness', 
                                                          num_trees=2, max_features=5)
clf_rf
```

```python
# Ensemble AdaBoost
clf_adab, y_test, y_predict = StartModSKL.classification_adab(cls_data, dependent_label='Happiness', num_trees=2)
clf_adab
```

```python
# Ensemble Stochastic Gradient Boosting
clf_sgb, y_test, y_predict = StartModSKL.classification_sgb(cls_data, dependent_label='Happiness', num_trees=3)
clf_sgb
```
