### Question: How would you describe the trajectories of marriages, divorces, newborns,
### and deaths in the years 1950-2015?

```python
# setup absolute path to location of package Starts and config-file 
from inspect import getsourcefile
import os.path as path, sys
current_dir = path.dirname(path.abspath(getsourcefile(lambda:0)))
sys.path.insert(0, current_dir[:current_dir.rfind(path.sep)])

# import package Starts
from Starts.startvis import *
%matplotlib inline
rcParams['figure.figsize'] = 15, 6
marriages_1950_2015 = pd.read_csv('./data/marriages_1950_2015.csv')
newborn_1950_2015 = pd.read_csv('./data/newborn_1950_2015.csv')
divorces_1950_2015 = pd.read_csv('./data/divorces_1950_2015.csv')
death_1950_2015 = pd.read_csv('./data/death_1950_2015.csv')
```

```python
len(marriages_1950_2015), len(newborn_1950_2015), len(divorces_1950_2015), len(death_1950_2015)
```

```python
marriages_1950_2015.columns, newborn_1950_2015.columns, divorces_1950_2015.columns, death_1950_2015.columns
```

```python
# set column 'Jahr' as index
marriages_1950_2015.set_index('Jahr', inplace=True)
newborn_1950_2015.set_index('Jahr', inplace=True)
divorces_1950_2015.set_index('Jahr', inplace=True)
death_1950_2015.set_index('Jahr', inplace=True)
```

```python
marriages_1950_2015.plot()
plt.title("Marriages trajectory")

newborn_1950_2015.plot()
plt.title("Newborn trajectory")

divorces_1950_2015.plot()
plt.title("Divorces trajectory")

death_1950_2015.plot()

plt.legend()
plt.title("Deaths trajectory")
```

```python
marriages_divorces_1950_2015 = pd.concat([marriages_1950_2015, divorces_1950_2015],axis =1)
```

```python
marriages_divorces_1950_2015.plot()
```

```python
newborn_1950_2015.rename(columns={'männlich': 'männlich_newborn', 'weiblich':'weiblich_newborn', 'Insgesamt': 'Insgesamt_newborn'}, inplace=True)
death_1950_2015.rename(columns={'männlich': 'männlich_death', 'weiblich':'weiblich_death', 'Insgesamt': 'Insgesamt_death'}, inplace=True)
```

```python
newborn_1950_2015.head()
```

```python
death_1950_2015.head()
```

```python
newborn_death_1950_2015 = pd.concat([newborn_1950_2015, death_1950_2015], axis=1)
newborn_death_1950_2015.plot()
plt.title("Comparing between Newborn and death from 1950 to 2015")
```

```python
marriages_divorces_1950_2015.to_csv('./data/marriages_divorces_1950_2015.csv', index=True)
newborn_death_1950_2015.to_csv('./data/newborn_death_1950_2015.csv', index=True)
```
