```python
import pandas as pd
import numpy as np
# from scipy.sparse import csr_matrix
```

```python
feature_set = pd.read_csv('dataset/features-400.txt', sep='\s+', names=['Email_ID', 'Word_ID', 'Frequency'])
label_set = pd.read_csv('dataset/labels-400.txt', sep='\s+', names=['labels'])

number_of_columns = len(feature_set['Word_ID'])
# print (number_of_columns)
arr = np.zeros((len(label_set), number_of_columns), dtype=np.int)
```

```python
idr = 0  # identify row of matrix arr

for i in range(len(feature_set)):    
    j = feature_set['Email_ID'][i]  # comparing index for column of arr
    w = feature_set['Word_ID'][i] 
    f = feature_set['Frequency'][i]

    if (idr + 2 > j):                
        # sdf.at[j-1][w] = f
        # sdf.set_value(j-1, str(w), f)  # Warning: extrem slowly process
        arr[idr][i] = f
        # print (i, idr, idc, arr[idr][idc], j, w, f) 

    else:       
        idr = idr + 1            
        arr[idr][i] = f        
        # print (i, idr, idc, arr[idr][idc], j, w, f) 
        

# Convert array to sparse Matrix type DataFrame
# sp_arr = csr_matrix(arr)
# sdf2 = pd.SparseDataFrame(sp_arr)   # bad choice, it creates NaN numpy.float64, unable to go on.

# Convert to DataFrame
sdf = pd.DataFrame(arr)

# Update Header for sparse data frame matrix 'sdf'
header = list(sdf.columns.values)
new_header = feature_set['Word_ID']

# remove duplicate in header, Question: is it good choice ?
# new_header = new_header.drop_duplicates()
# print ("--------- ", len(new_header))

# replace header by word_id
sdf.rename(columns=dict(zip(header, new_header)), inplace=True)

# print (sdf.values[399])
sdf['Label'] = [l[0] for l in label_set.values]

r = [item for item in arr[1] if item > 0]
# r2 = [item for item in sdf.values[399] if item > 0]
# print (r)
# print (len(r))
# print (r2)
# print (len(r2))

sdf
```

```python
from sklearn.model_selection import train_test_split

# get all columns except the last one
X = sdf.iloc[:, :-1].values

# get last column as label
y = sdf.iloc[:, number_of_columns].values

X_train, X_test, y_train_target, y_test_target = train_test_split(X, y, test_size=0.3, random_state=0)

print (len(X_train))
print (len(X_test))
# print (X_train)
# y_train_target
```

** Testing with scikit_learn MultinomialNB **

```python
from sklearn.naive_bayes import MultinomialNB
clf = MultinomialNB()

clf.fit(X_train, y_train_target)
# predict with test_data
test_result = clf.predict(X_test)
print (test_result)
print ('\n')
print (y_test_target)
z = test_result == y_test_target
z
accuracy_test = ((z == True).sum() / len(z)) * 100
accuracy_test
```
