```python
countries = ['China', 'Germany', 'Italy', 'Japan', 'Australia', 'Luxembourg', 'US', 'United Kingdom', 'Canada', 'Vietnam', 'Philippines', 'Israel']

total_countries_vaccinated = total_confirmed_vaccine_global.groupby(by='Country_Region').sum()

# total_countries_vaccinated['vaccination_rate'] = total_countries_vaccinated['People_partially_vaccinated']/total_countries_vaccinated['Doses_admin']*100
total_countries_vaccinated['vaccination_speed']=total_countries_vaccinated.apply(lambda row: row.People_partially_vaccinated/row.Doses_admin*100, axis=1)
total_countries_vaccinated.sort_values(by='vaccination_speed', ascending=False)['vaccination_speed'].head(20).plot(kind='bar', title='Vaccination Speed', color='lightgreen')
plt.xticks(rotation=45)
total_countries_vaccinated.sort_values(by='vaccination_speed', ascending=False).head(20)
```

```python
total_countries_vaccinated.sort_values(by='vaccination_speed', ascending=False)['vaccination_speed'].tail(20).plot(kind='bar', title='Vaccination Speed', color='lightgreen')
plt.xticks(rotation=45)
total_countries_vaccinated.sort_values(by='vaccination_speed', ascending=False).tail(20)
```
