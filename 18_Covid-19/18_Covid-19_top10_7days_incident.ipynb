### 1. Top 10 countries with highest new CONFIRMED increasing average last 7 days

```python
plt.title("Top 10 countries with HIGHEST new CONFIRMED increasing last day")
countries_confirmed.iloc[-8:].diff().tail(7).mean().sort_values(ascending=False).head(10).plot(kind='bar', color=(0.8, 0.7, 0.6, 0.5), edgecolor='orange')
plt.xticks(rotation=0)
countries_confirmed.iloc[-8:].diff().tail(7).mean().sort_values(ascending=False).head(10)
```

```python
countries_high = countries_confirmed.iloc[-8:].diff().tail(7).mean()
print('Vietnam' in countries_high.loc[countries_high>300].sort_values(ascending=False))
countries_high.loc[countries_high>300].sort_values(ascending=True).plot(kind='bar', color='orange')
print("Total countries ", len(countries_high.loc[countries_high>300]))
```
