### 4. Top 10 countries with highest 7-days incidence

```python
total_7days_confirmed = countries_confirmed.iloc[-8:].diff().tail(7).sum()
total_7days_confirmed.sort_values(ascending=False).head(10)
```

```python
# check the incidence with certain ratio
ratio = 100000

total_7days_incidence = pd.DataFrame(columns=total_7days_confirmed.index)
total_7days_incidence.loc['Population'] = 0
total_7days_incidence.loc['7days_totalcases'] = 0

for country in country_pop_dict.keys():
    total_7days_incidence[country]['Population'] = country_pop_dict[country]
    total_7days_incidence[country]['7days_totalcases'] =  total_7days_confirmed[country]

total_7days_incidence.dropna(axis=1, inplace=True)
total_7days_incidence.loc['7days_incidence'] = total_7days_incidence.loc['7days_totalcases'].astype('float')/total_7days_incidence.loc['Population'].astype('float') * ratio
total_7days_incidence = total_7days_incidence.transpose().sort_values(by='7days_incidence', ascending=False)
total_7days_incidence
```

```python
plt.title("Total 7 days incidence confirmed")
total_7days_incidence.head(20)['7days_incidence'].plot(kind='bar', color='orange')
print("Vietnam among the countries with highest total 7 days incidence confirmed: ", 'Vietnam' in total_7days_incidence.head(50)['7days_incidence'])
print('Max incidence: ', total_7days_incidence.head(50)['7days_incidence'].max())
total_7days_incidence.head(20)['7days_incidence']
```

```python
print("The 7 days incidence confirmed in following countries\n")

for country in ['Vietnam', 'United Kingdom', 'Australia', 'Luxembourg', 'Germany', 'France', 'Belgium', 'US']:
    print(country, total_7days_incidence.loc[country], '\n')
```

```python
total_7days_fatal = countries_fatalities.iloc[-8:].diff().tail(7).sum()

total_7days_incidence_fatal = pd.DataFrame(columns=total_7days_fatal.index)
total_7days_incidence_fatal.loc['Population'] = 0
total_7days_incidence_fatal.loc['7days_totalcases'] = 0

for country in country_pop_dict.keys():
    total_7days_incidence_fatal[country]['Population'] = country_pop_dict[country]
    total_7days_incidence_fatal[country]['7days_totalcases'] =  total_7days_fatal[country]

total_7days_incidence_fatal.loc['7days_incidence'] = total_7days_incidence_fatal.loc['7days_totalcases'].astype('float')/ total_7days_incidence_fatal.loc['Population'].astype('float') * 100000    
total_7days_incidence_fatal = total_7days_incidence_fatal.transpose().sort_values(by='7days_incidence', ascending=False)
total_7days_incidence_fatal.dropna(inplace=True)
```

```python
plt.title("Total 7 days incidence fatal")
total_7days_incidence_fatal.head(50)['7days_incidence'].plot(kind='bar', color='red')
print("Vietnam among countries with highest total 7 days incidence fatal: ", 'Vietnam' in total_7days_incidence_fatal.head(50)['7days_incidence'])
'Max incidence: ', total_7days_incidence_fatal.head(50)['7days_incidence'].max()
```

```python
print("The 7 days incidence fatal in following countries\n")

for country in ['Vietnam', 'United Kingdom', 'Australia', 'Luxembourg', 'Germany', 'France', 'Belgium', 'US']:
    print(country, total_7days_incidence_fatal.loc[country], '\n')
```
