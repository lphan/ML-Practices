### 4. Top 10 countries with highest 7-days incidence

```python
total_7days_confirmed = countries_confirmed.iloc[-8:].diff().tail(7).sum()
total_7days_confirmed.sort_values(ascending=False).head(10)
```

```python
total_7days_incidence = pd.DataFrame(columns=total_7days_confirmed.index)
total_7days_incidence.loc['Population'] = 0
total_7days_incidence.loc['7days_totalcases'] = 0

for country in country_pop_dict.keys():
    total_7days_incidence[country]['Population'] = country_pop_dict[country]
    total_7days_incidence[country]['7days_totalcases'] =  total_7days_confirmed[country]

total_7days_incidence.loc['7days_incidence'] = total_7days_incidence.loc['7days_totalcases']/ total_7days_incidence.loc['Population'] * 100000    
total_7days_incidence = total_7days_incidence.transpose().sort_values(by='7days_incidence', ascending=False)
total_7days_incidence.dropna(inplace=True)
```

```python
total_7days_incidence
```

```python
plt.title("Total 7 days incidence confirmed")
total_7days_incidence.head(20)['7days_incidence'].plot(kind='bar', color='orange')
total_7days_incidence.head(20)['7days_incidence']
```

```python
for country in ['Australia', 'Luxembourg', 'Germany', 'Vietnam']: 
    print(country, total_7days_incidence.loc[country], '\n')
```
