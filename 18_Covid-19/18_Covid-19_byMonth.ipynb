```python
countries = ['China', 'Germany', 'Italy', 'Korea, South', 'Japan', 'Australia', 'Luxembourg', 'US', 'United Kingdom', 'Canada', 'Vietnam', 'Israel']

rcParams['figure.figsize'] = 20, 20
countries_confirmed_month[countries].plot(subplots=True, legend=True, color='orange')

plt.xticks(np.arange(0, len(countries_confirmed_month), 1))
plt.xlabel('Months')
plt.ylabel('Infected cases')
plt.legend()
countries_confirmed_month[countries]
```

```python
rcParams['figure.figsize'] = 20, 10
countries_confirmed_month.apply(lambda x: x.diff()).fillna(0).sum(axis=1).plot(kind='bar', color='orange')
plt.xticks(rotation=0)
plt.title("Confirmed Cases every month in the World")
countries_confirmed_month.apply(lambda x: x.diff()).fillna(0).sum(axis=1)
```

```python
rcParams['figure.figsize'] = 20, 20
countries_fatalities_month[countries].plot(subplots=True, legend=True, color='red')

plt.xticks(np.arange(0, len(countries_fatalities_month), 1))
plt.xlabel('Months')
plt.ylabel('Death cases')
plt.legend()
countries_fatalities_month[countries]
```

```python
rcParams['figure.figsize'] = 20, 10

countries_fatalities_month.apply(lambda x: x.diff()).fillna(0).sum(axis=1).plot(kind='bar', color='red')
plt.xticks(rotation=0)
plt.title("Fatalities every month in the World")
countries_fatalities_month.apply(lambda x: x.diff()).fillna(0).sum(axis=1)
```

```python
# Confirmed by month
countries = ['Germany', 'Luxembourg', 'Vietnam']
rcParams['figure.figsize'] = 20, 10

countries_confirmed_month.apply(lambda x: x.diff()).fillna(0)[countries].plot(legend=True, kind='bar')

plt.xticks(np.arange(0, len(countries_confirmed_month), 1), rotation=0)
plt.xlabel('Months')
plt.ylabel('Infected cases')
plt.title("CONFIRMED BY MONTH")
plt.legend()
countries_confirmed_month.apply(lambda x: x.diff()).fillna(0)[countries]
```

```python
# Confirmed by month check on-demand group
countries = ['Thailand', 'Cambodia', 'Laos', 'Burma', 'Malaysia', 'Indonesia', 'Singapore', 'Vietnam']
rcParams['figure.figsize'] = 20, 10

countries_confirmed_month.apply(lambda x: x.diff()).fillna(0)[countries].plot(legend=True, kind='bar')

plt.xticks(np.arange(0, len(countries_confirmed_month), 1), rotation=0)
plt.xlabel('Months')
plt.ylabel('Infected cases')
plt.title("CONFIRMED BY MONTH in Group")
plt.legend()
countries_confirmed_month.apply(lambda x: x.diff()).fillna(0)[countries]
```

```python
# Fatalities by month
countries = ['Germany', 'Luxembourg', 'US', 'Vietnam', 'Israel']

countries_fatalities_month.apply(lambda x: x.diff()).fillna(0)[countries].plot(legend=True, kind='bar')

plt.xticks(np.arange(0, len(countries_confirmed_month), 1), rotation=0)
plt.xlabel('Months')
plt.ylabel('Fatalities cases')
plt.title("FATALITIES BY MONTH")
plt.legend()
countries_fatalities_month.apply(lambda x: x.diff()).fillna(0)[countries]
```
